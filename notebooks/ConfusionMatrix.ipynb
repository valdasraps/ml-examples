```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import confusion_matrix, roc_curve, roc_auc_score, precision_recall_curve
from sklearn import datasets

import numpy as np
import matplotlib.pyplot as plt
```

## Dataset

```python
wine = datasets.load_wine()
x = wine.data[:,[0,1]]
y = wine.target
```

```python
y[y == 2] = 1
```

```python
plt.scatter(x[:,0][y == 0.], x[:,1][y == 0.])
plt.scatter(x[:,0][y == 1.], x[:,1][y == 1.])
plt.show()
```

## Sklearn LogisticRegression

```python
log = LogisticRegression(solver='lbfgs')
log.fit(x,y)
```

```python
w0 = log.intercept_[0]
wS = log.coef_[0]

print(f'intercept = {w0}')
print(f'slopes    = {wS}')
```

```python
x1 = np.linspace(np.min(x[:,0]), np.max(x[:,0]), num = 10)
y1 = -( wS[0] * x1 + w0)/wS[1]
plt.scatter(x[:,0][y == 0.], x[:,1][y == 0.])
plt.scatter(x[:,0][y == 1.], x[:,1][y == 1.])
plt.plot(x1, y1, '-r')
plt.ylim((np.min(x[:,1]), np.max(x[:,1])))
plt.show()
```

```python
p = log.predict_proba(x)
```

## Confusion matrix (sklearn)

```python
confusion_matrix(y, log.predict(x))
```

```python
roc = roc_curve(y, p[:,1])
plt.plot(roc[0], roc[1])
plt.show()
```

```python
roc_auc_score(y,p[:,1])
```

```python
prc = precision_recall_curve(y,p[:,1])
plt.plot(prc[1], prc[0])
plt.show()
```

## Confusion matrix

```python
class CM:
    
    def __init__(self, y, p, th = 0.5):
        
        self.n = len(y)
        
        self.c = np.zeros(y.shape, dtype=int)
        self.c[p[:,1] >= th] = 1
        
        self.tp = np.sum((y == 1) & (self.c == 1))
        self.fp = np.sum((y == 0) & (self.c == 1))
        self.fn = np.sum((y == 1) & (self.c == 0))
        self.tn = np.sum((y == 0) & (self.c == 0))
        
    def acc(self):
        return (self.tp + self.tn) / self.n
        
    def err(self):
        return (self.fp + self.fn) / self.n

    def precision(self):
        return self.tp / (self.tp + self.fp)

    def recall(self):
        return self.tp / (self.tp + self.fn)

    def sensitivity(self):
        return self.recall()

    def null_err(self):
        p = self.tp + self.fn
        n = self.tn + self.fp
        if p < n: return p / self.n
        else: return n / self.n

    def specificity(self):
        return self.tn / (self.tn + self.fp)

    def fpr(self):
        return 1 - self.specificity()

    def tpr(self):
        return self.recall()

    def cm(self):
        return [[ self.tp, self.fp ],
                [ self.fn, self.tn ]]
    
```

```python
CM(y, p, 0.5).cm()
```

```python
roc = []
prc = []

for i in np.arange(0,1.1,.005):
    cm = CM(y, p, i)
    roc.append([cm.fpr(),cm.tpr()])
    prc.append([cm.recall(),cm.precision()])

roc = np.array(roc)
prc = np.array(prc)
```
