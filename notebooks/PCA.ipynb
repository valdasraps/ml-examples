```python
from sklearn import datasets
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler

from IPython.display import Image  

import numpy as np
import matplotlib.pyplot as plt
```

## Dataset

```python
wine = datasets.load_wine()
x = wine.data
x.shape
```

```python
plt.scatter(x[:,0], x[:,1])
plt.show()
```

## Sklearn PCA

```python
d = 13
pca = PCA(d)
pca.fit(x)
```

```python
plt.bar(range(1,d+1), pca.explained_variance_ratio_)
plt.yscale('log')
plt.show()
```

```python
xp = pca.transform(x)
```

```python
plt.scatter(xp[:,0], xp[:,1])
plt.show()
```

## PCA step-by-step

```python
x1 = StandardScaler().fit_transform(x)
```

```python
plt.scatter(x[:,0], x[:,1])
plt.scatter(x1[:,0], x1[:,1])
plt.show()
```

```python
cov = np.cov(x1.T)
cov.shape
```

```python
L, C = np.linalg.eig(cov)
print(L.shape,C.shape)
```

```python
plt.bar(range(1,d+1), L)
#plt.yscale('log')
plt.show()
```

```python
L = L[np.flip(np.argsort(L))]
C = C[np.flip(np.argsort(L))]
```

```python
plt.bar(range(1,d+1), L)
#plt.yscale('log')
plt.show()
```

```python
plt.bar(range(1,d+1), L / np.sum(L))
plt.yscale('log')
plt.show()
```

```python
W = C[:,:2]
W.shape
```

```python
xp = x1.dot(W)
plt.scatter(xp[:,0], xp[:,1])
plt.show()
```
