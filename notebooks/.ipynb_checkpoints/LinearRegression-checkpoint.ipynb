```python
from sklearn.linear_model import LinearRegression
from sklearn import datasets
import numpy as np
import matplotlib.pyplot as plt
```

## Dataset

```python
boston = datasets.load_boston()
x = boston.data[:,5][:,np.newaxis]
y = boston.target[:,np.newaxis]
```

```python
plt.scatter(x,y)
plt.show()
```

## Sklearn LinearRegression

```python
reg = LinearRegression()
reg.fit(x,y)
```

```python
w1 = reg.coef_[0][0]
w0 = reg.intercept_[0]
print(f'intercept = {w0}')
print(f'slope     = {w1}')
```

```python
x1 = np.linspace(np.min(x), np.max(x), num = 10)
y1 = w1 * x1 + w0
plt.scatter(x,y)
plt.plot(x1, y1, '-r')
plt.show()
```

## Closed solution opt#1

$$w_1=\frac{\sum_{i=1}^n{(x_i-\bar{x})}{(y_i-\bar{y})}}{\sum_{i=1}^n{(x_i-\bar{x})}^2}$$

$$w_0=\bar{y}-w_1\bar{x}$$

```python
w1 = np.sum((x - np.mean(x)) * (y - np.mean(y))) / np.sum((x - np.mean(x)) * (x - np.mean(x)))
w0 = np.mean(y) - w1 * np.mean(x)
print(f'intercept = {w0}')
print(f'slope     = {w1}')
```

```python
x1 = np.linspace(np.min(x), np.max(x), num = 10)
y1 = w1 * x1 + w0
plt.scatter(x,y)
plt.plot(x1, y1, '-r')
plt.show()
```

## Closed solution opt#2

$$w_1=\frac{\sum_i^n{x_{i}y_{i}}-\frac{\sum_i^n{x_{i}}\sum_i^n{y_{i}}}{n}}
{\sum_i^n{x_{i}^2}-\frac{\sum_i^n{x_{i}}\sum_i^n{x_{i}}}{n}}$$

$$w_0=\bar{y}-w_1\bar{x}$$

```python
n = len(x)
w1 = (np.sum(x * y) - (np.sum(x) * np.sum(y))/n)/(np.sum(x*x) - (np.sum(x) * np.sum(x))/n)
w0 = np.mean(y) - w1 * np.mean(x)
print(f'intercept = {w0}')
print(f'slope     = {w1}')
```

```python
x1 = np.linspace(np.min(x), np.max(x), num = 10)
y1 = w1 * x1 + w0
plt.scatter(x,y)
plt.plot(x1, y1, '-r')
plt.show()
```
